```python
import pandas as pd

dialogs_2_data = pd.read_csv("data/DATA/dialogs_2.csv")
```

```python
dialogs_2_data = dialogs_2_data.dropna(subset=['client_id','user_id'], how='all')
dialogs_2_data['user_id'] = dialogs_2_data['user_id'].fillna(-1)
dialogs_2_data.head()
```

```python
from datetime import datetime

def parseDatetime(dt_str):
    y = int(dt_str[0:4])
    m = int(dt_str[5:7])
    d = int(dt_str[8:10])
    
    H = int(dt_str[11:13])
    M = int(dt_str[14:16])
    S = int(dt_str[17:19])
    dt = datetime(y, m, d, H, M, S)
    return dt

dialogs_2_data["dt"] = dialogs_2_data["datetime"].apply(lambda x: parseDatetime(x))
print(dialogs_2_data.head())
```

```python
import numpy as np
import matplotlib.pyplot as plt


def userAnswerTime(data):
    data = data.sort_values(by=['datetime'])
    delta = [0]
    values = data['dt'].get_values()
    for i in range(0,len(data)-1):
        val = (values[i+1] - values[i])/np.timedelta64(1, 's')
        if (val < 60*10):
            delta.append(val)
        else:
            delta.append(0)
    data['user_answer_time'] = delta
    return data

group = dialogs_2_data.groupby("user_id")
group = group.apply(userAnswerTime)
```

```python
dialogs_2_data = group.drop("user_id",1).reset_index()
```
